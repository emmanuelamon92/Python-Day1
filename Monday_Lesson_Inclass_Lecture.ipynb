```python
num1 = 2
num2 = 5.4

result = num1 + num2

print(result)
type(result)

# Short-hand version
result += 2 # result = result + 2
print(result)
```

```python
result_diff = num2 - num1

print(result_diff)

# Short-hand version
result_diff -= 1
print(result_diff)
print("Type", type(result_diff))

result_diff = int(result_diff)
print(result_diff)
print("Type", type(result_diff))
```

### If Statements

```python
# Available operators: Greater(>), Less(<),Equal(==)
# Greater or Equal(>=), Less or Equal (<=)

# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F

num1 = 15
num2 = 10

if num2 == num1:
    print("Equal Values")
else:
    print("Not Equal")
    
if num2 > num1:
    print("Num2 is greater")
elif num1 > num2:
    print("Num1 is less")
```

### Lists

```python
l_1 = [] # Empty List Declaration

names= ['Max', 'Cindy', 'Kathy', 'Bob', 'Nate']

print(names)
```

##### Indexing a List

```python
# [start:stop:step]

#Siinge Index
print(names[-1])

# Print starting at index 1 going to the end
print(names[1:])

# Print stgarting at the beginning of the List to N (N = the number we want to stop at)
print(names[:2])

# Print starting at index 1 and skip every other on to the end
print(names[0::2])

# Print starting at the back and reverse order
print(names[::-1])
```

##### Lists Within Lists

```python
# Lists are able to hold any type of element, including other lists themselves
# They cna go as deep as you wnat, having many lists nested inside of each other

names = ['max', 'sam', 'cindy', ['bob', 'nate', 'brian']]
print(names)

names[3].append('julie')
print(names)
print(names[3][2])
```

##### Looping Through Lists

```python
# two ways to Loop through a List, one is by index, the other is by using the 'in' keyword

# Lop with a for Loop (By Index)
for i in range(len(names)):
    print(i, names[i])
    
print("")
# Loop with a for Loop (Using the 'in' keyword)
for i in names:
    if isinstance(i, list):
        for j in i:
            print(j)
        else:
            print(i)
            
```
